```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def ma(X, w=None):
    if(not w): w = len(X)
    avgs = []
    for i in range(len(X)):
        numer = np.sum(X[0:i+1])
        den = i+1
        avgs.append(numer/den)
    return np.array(avgs)
```

## First Plot

```python
fig, axs = plt.subplots(1, 2)
fig.set_size_inches(40, 12.5)
fig.tight_layout()
```

```python
dpp_hits1 = pd.read_csv("DPP_Experiment/client_1/hit_rate.csv")
# dpp_hits2 = pd.read_csv("DPP_Experiment/client_2/hit_rate.csv")
# dpp_hits3 = pd.read_csv("DPP_Experiment/client_3/hit_rate.csv")
global_hits1 = pd.read_csv("Experiment_Clients/client_1/hit_rate.csv")
# global_hits2 = pd.read_csv("Experiment_Clients/client_2/hit_rate.csv")
# global_hits3 = pd.read_csv("Experiment_Clients/client_3/hit_rate.csv")

plt.figure(figsize=(10*1.5, 7*1.5))
plt.xlabel("Timeslot", fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.ylabel("Average Cache Hit", fontsize=30)
plt.xlim(left=0, right=1000)
plt.ylim(bottom=0, top=1.0)
plt.plot(ma(dpp_hits1), label="Client 1", marker='o', markevery=50, markersize=15, c='orange', linewidth=2)
# plt.plot(ma(dpp_hits2), label="Client 2", marker='d', markevery=50, markersize=15, c='blue', linewidth=2)
# plt.plot(ma(dpp_hits3), label="Client 3", marker='^', markevery=50, markersize=15, c='purple', linewidth=2)
plt.plot(ma(global_hits1), label="Global 1", marker='x', markevery=50, markersize=15, c='orange', linewidth=2)
# plt.plot(ma(global_hits2), label="Global 2", marker='x', markevery=50, markersize=15, c='blue', linewidth=2)
# plt.plot(ma(global_hits3), label="Global 3", marker='x', markevery=50, markersize=15, c='purple', linewidth=2)
plt.legend(loc='lower right', prop={'size': 20})
plt.grid()
plt.savefig("cache_hit.pdf", bbox_inches='tight')
```
